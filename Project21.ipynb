### 2.2 Arithmetic Operations

### 2.2.1 Pointwise Addition

```python
A = cv2.imread('cameraman.jpg')

def f(x):
    return x + 40

def point_operation(image, method):
    for row in range(0, image.shape[0]):
        for col in range(0, image.shape[1]):
            temp = method(image[row][col][0])
            if temp > 255: 
               image[row][col] = 255
            elif temp < 0:
               image[row][col] = 0
            else:
               image[row][col] = temp
    return image


image_modified = point_operation(A.copy(), f)


plt.figure(figsize=(20,20))
plt.subplot(221)
plt.imshow(A)
plt.title('Color')
plt.subplot(222)
plt.imshow(image_modified)
plt.title('Color')
```

```python
histr = cv2.calcHist([A],  
                     [0], None, 
                     [256],  
                     [0, 256]) 
      
plt.plot(histr) 

# Limit X - axis to 256 
plt.xlim([0, 256]) 
      
plt.show()
plt.show()
```

### 2.3 Linear Contrast Stretching

```python
A_path = "testpic.jpg"
img = cv2.imread(A_path)

# normalize float versions
norm_img1 = cv2.normalize(img, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
norm_img2 = cv2.normalize(img, None, alpha=0, beta=1.2, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)

# scale to uint8
norm_img1 = (255*norm_img1).astype(np.uint8)
norm_img2 = np.clip(norm_img2, 0, 1)
norm_img2 = (255*norm_img2).astype(np.uint8)

# write normalized output images
cv2.imwrite("cameraman_normalize1.jpg",norm_img1)
cv2.imwrite("cameraman_normalize2.jpg",norm_img2)

plt.figure(figsize=(20,20))
plt.subplot(221)
plt.imshow(img)
plt.title('Color')
plt.subplot(222)
plt.imshow(norm_img1)
plt.title('Color')
plt.subplot(223)
plt.imshow(norm_img2)
plt.title('Color')
```

### 2.4 Density Slicing

```python
# Load the image
img = cv2.imread('cameraman.jpg',0)
# Find width and height of image
row, column = img.shape
# Create an zeros array to store the sliced image
img1 = np.zeros((row,column),dtype = 'uint8')
 
# Specify the min and max range
min_range = 10
max_range = 60
 
# Loop over the input image and if pixel value lies in desired range set it to 255 otherwise set it to 0.
for i in range(row):
    for j in range(column):
        if img[i,j]>min_range and img[i,j]<max_range:
            img1[i,j] = 255
        else:
            img1[i,j] = 0

imgplot = plt.imshow(img1)
plt.show()
```

```python
histr = cv2.calcHist([img1],  
                     [0], None, 
                     [256],  
                     [0, 256]) 
      
plt.plot(histr) 

# Limit X - axis to 256 
plt.xlim([0, 256]) 
plt.ylim([0, 1600]) 
      
plt.show()
plt.show()
```
